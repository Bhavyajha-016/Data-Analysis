### ............................DATA DICTIONARY .................................

```python
# Data Dictionary
"""
-game_id: Unique identifier for each game.

-competition_id_x: Identifier for the competition in which the game was played.

-season: The season during which the game took place.

-round: The round or matchday of the game within the season.

-date_x: The date and time when the game was played.

-home_club_goals: Number of goals scored by the home club.

-away_club_goals: Number of goals scored by the away club.

-home_club_position: The league position of the home club before the game.

-away_club_position: The league position of the away club before the game.

-home_club_manager_name: Name of the home club's manager.

-away_club_manager_name: Name of the away club's manager.

-stadium: Name of the stadium where the game was played.

-attendance: Number of spectators attending the game.

-referee: Name of the referee officiating the game.

-home_club_name: Name of the home club.

-away_club_name: Name of the away club.

-aggregate: Aggregate score (if part of a two-legged tie).

-competition_type: Type of competition (e.g., domestic league, cup competition).

-game_event_id: Unique identifier for each game event.

-date_y: Date and time of the game event.

-minute: The minute during the game when the event occurred.

-type_y: Type of game event (e.g., goal, substitution, card).

-description: Description of the game event.

-appearance_id: Unique identifier for the player's appearance in the game.

-player_id: Unique identifier for the player.

-date: Date when the player information was recorded.

-player_name_y: Name of the player involved in the event.

-competition_id_y: Identifier for the competition related to the event.

-yellow_cards: Number of yellow cards received by the player.

-red_cards: Number of red cards received by the player.

-goals: Number of goals scored by the player in the game.

-assists: Number of assists made by the player in the game.

-minutes_played: Number of minutes the player played in the game.

-name: Name of the player.

-last_season: Last season the player played.

-current_club_id: Identifier for the player's current club.

-player_code: Code or unique identifier for the player.

-country_of_birth: Country where the player was born.

-date_of_birth: Player's date of birth.

-sub_position: Specific position the player played in the game.

-position_y: General position category (e.g., Defender, Midfielder).

-foot: Preferred foot of the player (e.g., right, left, both).

-height_in_cm: Height of the player in centimeters.

-market_value_in_eur: Market value of the player in Euros.

-highest_market_value_in_eur: Highest market value of the player in Euros throughout their career.

-contract_expiration_date: Date when the player's current contract expires.

-agent_name: Name of the player's agent.

"""
```

### ........................... DATA VISUALIZATION ..............................

### There are some Focus Areas which I have consider to do the analysis.

### 1) Performance Analysis

### - Evaluating players performace based on Assist(An assist is a significant statistic used to measure a player's contribution to a team's goal-scoring efforts).

```python
merge_df.head()
```

```python
P_df =  merge_df[merge_df.type_y == 'Goals']
perform=P_df.groupby(by = 'player_name_y')['assists'].sum().sort_values(ascending = False) 
df_1 =perform[: 18]
df_1
```

```python
plt.figure(figsize = (7,5))
ax= sns.barplot(x=df_1.index, y=df_1.values)
plt.xticks(rotation = 90)
ax.bar_label(ax.containers[0], fontsize=10)
plt.title("Players Who Goals high while Assisting")
plt.ylabel('Total Assist Count')
plt.show()
```

###  Aron Johannsson, Christian Pulisic, Jozy Altidore are the players who secure goal while assisting the team.

### 2) Player's profile & Market Value

### - How do player's height impact the match?

### - How do player's age affect the market value?

```python
h= P_df.groupby(by = 'height_in_cm')['player_name_y'].count().sort_values()
h
```

```python
plt.figure(figsize = (7,5))
ax= sns.barplot(x=h.index, y=h.values)
plt.xticks(rotation = 90)
ax.bar_label(ax.containers[0], fontsize=10)
plt.title("Players height")
plt.ylabel('Height')
plt.show()
```

```python
sns.histplot(P_df.height_in_cm, bins =10, color="orange")
plt.xlabel("Players Height")
plt.ylabel("Height Value")
plt.show()
```

### Using Histogram & Bar Graph I can say Highest number of players belongs to height group 177-185.

```python
market= P_df.groupby(by = ['date_of_birth'])['market_value_in_eur'].sum().sort_values(ascending = False)
market
```

```python
plt.figure(figsize= (9,5))
sns.lineplot(x=market.index, y = market.values)

plt.title("Market Value of Players based on Age")
plt.ylabel('')
plt.show()
```

### By line Graph I came to know that Players who born on 1988 have high market value. Highest market value are holded by thoes players who are young.

### 3) Stadium Analysis 

```python
stadium= P_df.groupby(by = 'stadium')['game_event_id'].count().sort_values(ascending = False)
top_15_stadium= stadium[: 16]
```

```python
plt.figure(figsize = (7,5))
ax= sns.barplot(x=top_15_stadium.index, y=top_15_stadium.values)
plt.xticks(rotation = 90)
ax.bar_label(ax.containers[0], fontsize=10)
plt.title("Stadium Analysis")
plt.xlabel("Top 15 Stadiums")
plt.ylabel('Total Events')
plt.show()
```

### Highest number of match events takes place in stadium AFAS Stadium. 

### 4) Referees Analysis

### - Which referees officiated highest number of matches?

```python
referee= P_df.groupby(by = 'referee')['game_id'].count().sort_values(ascending = True)
referee
top_10_referee= referee.tail(10)
top_10_referee
```

```python
plt.barh(top_10_referee.index, top_10_referee.values, color="darkblue")
plt.xlabel('Total Game Id')
plt.ylabel('Top 10 Referee')
plt.title("Referee's Analysis")      
plt.show() 
```

### These are the top 10 referees list who officiated the match. By this we can say Marco Fritz is one of the best referees in terms of fair decison making.

### 5) Substitution Pattern

### - Which players plays well in substitution? (Substitution in football refers to the process of replacing one player on the field with another player.Substitutions can be made for various reasons, including tactical changes, injuries, or to give players rest)

```python
S_df =  merge_df[merge_df.type_y == 'Substitutions']
Substitute= S_df.groupby(by = 'player_name_y')['goals'].sum().sort_values(ascending = True)
top_10 = Substitute.tail(10)
top_10
```

```python
plt.barh(top_10.index, top_10.values, color="green")
plt.xlabel('Total Goals')
plt.ylabel('Top 10 Players')
plt.title("Substitution Analysis")      
plt.show() 
```

### After substitution Aron Johannsson, Christian Pulisic & Bobby Wood Goals are the highest.

### 6) Event Analysis

### - In which years high number of event occurs?

```python
merge_df['year'] = merge_df['date'].dt.year
merge_df.shape
```

```python
event= merge_df.groupby(by = ['year'])['game_event_id'].count()
event
```

```python
plt.figure(figsize= (9,5))
plt.plot(event.index, event.values, color= 'red', marker='o', markerfacecolor='blue')
sns.regplot(x=event.index, y=event.values, line_kws={'linestyle': '--', 'linewidth': 2})
plt.title("Event Analysis")
plt.ylabel('Count of Events')
plt.show()
```

### Highest game events takes place in year 2013. As we go forward trendline is declining over the period.

### 7) Contract Management 

### - In which year highest number of contract get expires? 

```python
merge_df['contract_expiration_year'] = merge_df['contract_expiration_date'].dt.year
merge_df.shape
```

```python
contract_expiration= merge_df.groupby(by = ['contract_expiration_year'])['player_name_y'].count()
contract_expiration
```

### Team Comparison

### Home Club Team - The home club is the team that hosts the match at their own stadium.
### Away Club Team - The away club is the team that travels to the opponent’s stadium to play the match.

### - Analyse the team on the basis of player's performance.

```python
H_club= P_df.groupby(by = 'player_name_y')['home_club_goals'].sum().sort_values()
Home = H_club.tail(10)
Home
```

```python
A_club= P_df.groupby(by = 'player_name_y')['away_club_goals'].sum().sort_values()
Away = A_club.tail(10)
Away
```

```python
plt.figure(figsize= (9,5))
plt.plot(Home.index, Home.values, color= 'red', marker='o', markerfacecolor='blue')
plt.plot(Away.values, color= 'green', marker='o', markerfacecolor='yellow')
plt.xticks(rotation = 90)
plt.title("Team Comparison")
plt.ylabel('Total Goals')
plt.xlabel("RED - Home Club Team Goals / Green - Away Club Team Goals ")
plt.show()
```

```python
print("Total Home Club Goals : ",P_df['home_club_goals'].sum())
print("Total Away Club Goals : ",P_df['away_club_goals'].sum())
```

### Total goals secure by 'home club team' are high as comparison to 'away club team'. Top 5 Players who plays best are Aron Johannsson, Christian Pulisic, Jozy Altidore, Fabian Johnson, Timothy Chandler.
